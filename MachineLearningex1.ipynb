Making up of Gradient Descent and Linear Regression Model

```python
df['feature'] = 1
y = np.array(df[['profit']])
X = np.array(df[['feature','population']])
theta = np.array([1,2]) # Initializing a dynamic Theta to begin with
m = X.shape[0]
alpha = 0.001
```

```python
def Cost(X,y,Theta,m):
    return ((X@Theta  - np.transpose(y))**2 /(2*m)).sum()
```

```python
def UpdatedTheta(X,y,Theta,m,alpha):
    Theta0 = Theta[0]
    Theta1 = Theta[1]
    Theta0 = Theta0 -alpha*((X@Theta - np.transpose(y))/m*X[:,0]).sum()
    Theta1 = Theta1 -alpha*((X@Theta - np.transpose(y))/m*X[:,1]).sum()
    return np.array([Theta0,Theta1])
```

```python
epoch =0
theta = np.array([0,0])
alpha = 0.001
costdata = []
while epoch<1500:
    epoch+=1
    costdata.append(Cost(X,y,theta,m))
    d = pd.DataFrame(costdata).reset_index()
    plt.plot(d[0],d['index'],color='r',marker='.')
    plt.xlabel('Iteration')
    plt.ylabel('Cost')
    plt.show
    theta = UpdatedTheta(X,y,theta,m,alpha)
print('Final Theta value is:',theta)
```

```python
def projectedValue(X,Theta):
    return X@Theta
```

```python
df['predicted']=df[['profit','population']].apply(lambda x: projectedValue(np.array([1,x.population]),theta),axis = 1)
```

```python
plt.scatter(df.population,df.profit,c='r',marker='x')
plt.plot(df.population,df.predicted,c='b',marker='_')
plt.xlabel('Population')
plt.ylabel('Profit')
plt.show()
```
